```python
import os
import json
import pandas as pd
from os import walk

def process_json_to_df(file_path):
    try:
        # Read JSON file
        with open(file_path, 'r') as file:
            data = json.load(file)
        
        abstracts_data = data.get('abstracts-retrieval-response', {})
        coredata = abstracts_data.get('coredata', {})
        authors_data = abstracts_data.get('authors', {}).get('author', [])
        subject_areas = abstracts_data.get('subject-areas', {}).get('subject-area', [])
        

        # Helper function to get author count
        def get_sub_type():
            return coredata.get('subtype')
        def get_author_count():
            try:
                return len(authors_data) if authors_data else 0
            except Exception:
                return 0

        # Helper function to get subject area count
        def get_subject_area_count():
            try:
                return len(subject_areas) if subject_areas else 0
            except Exception:
                return 0

        def get_ref_count():
            try:
                bibliography = abstracts_data.get('item',{}).get('bibrecord', {}).get('tail', {}).get('bibliography', {})
                return bibliography.get('@refcount')
            except Exception:
                return None
        def get_unique_affiliations():
            try:
                affiliations = []
                for author in authors_data:
                    affiliation = author.get('affiliation')
                    if isinstance(affiliation, list):
                        for aff in affiliation:
                            affiliations.append(aff.get("@id"))
                    elif isinstance(affiliation, dict):
                        affiliations.append(affiliation.get("@id"))
                
                # Remove duplicates using set and count unique affiliations
                unique_affiliations = len(set(filter(None, affiliations)))
                return unique_affiliations
            except Exception as e:
                print(f"Error in get_unique_affiliations: {str(e)}")
                return 0

        # Helper function to get citation count
        def cite_count():
            return coredata.get('citedby-count')
        def get_open_access():
            try:
                return coredata.get('openaccess', None)  # Assumes 'openaccess' exists in coredata
            except Exception:
                return None
        
        # Create a dictionary with the data
        paper_data = {
            'Title': coredata.get('dc:title') or None,
            'ref_count': get_ref_count(),
            'Author_count': get_author_count(),
            'Subject_area_count': get_subject_area_count(),
            'Unique_affiliations_count': get_unique_affiliations(),
            'Open_access': get_open_access(),
            'subtype': get_sub_type(),
            'Cite_count': cite_count()
        }
        
        return paper_data
    
    except Exception as e:
        print(f"Error processing file {file_path}: {str(e)}")
        # Return empty dictionary with None values
        return {
            'Title': None,
            'Author_count': None,
            'Subject_area_count': None,
            'subtype': None,    
            'Cite_count': None
        }

# Directory path and processing
base_path = r'C:\Users\USER\Desktop\DSDE_project\Project\2023\2023'
all_data = []
f = []
for (dirpath, dirnames, filenames) in walk(r'C:\Users\USER\Desktop\DSDE_project\Project\2023'):
    f.extend(filenames)
    break
for i in range(len(f)):
# for i in range(1):
    file_number = str(i).zfill(5)
    file_path = base_path + file_number
    
    if os.path.exists(file_path):
        paper_data = process_json_to_df(file_path)
        all_data.append(paper_data)
    
    if i % 100 == 0:
        print(f"Processed {i} files...")

df = pd.DataFrame(all_data)

df.to_csv('2023_ML_V3.csv',index=False)
```
