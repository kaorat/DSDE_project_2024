```python
import pandas as pd
import os
```

```python
script_dir = os.getcwd()
file_path_2018 = os.path.join(script_dir, '2018_trend_topic.csv')
file_path_2019 = os.path.join(script_dir, '2019_trend_topic.csv')
file_path_2020 = os.path.join(script_dir, '2020_trend_topic.csv')
file_path_2021 = os.path.join(script_dir, '2021_trend_topic.csv')
file_path_2022 = os.path.join(script_dir, '2022_trend_topic.csv')
file_path_2023 = os.path.join(script_dir, '2023_trend_topic.csv')
data_2018 = pd.read_csv(file_path_2018)
data_2019 = pd.read_csv(file_path_2019)
data_2020 = pd.read_csv(file_path_2020)
data_2021 = pd.read_csv(file_path_2021)
data_2022 = pd.read_csv(file_path_2022)
data_2023 = pd.read_csv(file_path_2023)

data_2018['Year'] = 2018
data_2019['Year'] = 2019
data_2020['Year'] = 2020
data_2021['Year'] = 2021
data_2022['Year'] = 2022
data_2023['Year'] = 2023
```

```python
# Merge 2018 and 2019

data_merged = pd.concat([data_2018, data_2019], axis=0)

# Merge merged and 2020

data_merged = pd.concat([data_merged, data_2020], axis=0)

# Merge merged and 2021

data_merged = pd.concat([data_merged, data_2021], axis=0)

# Merge merged and 2022

data_merged = pd.concat([data_merged, data_2022], axis=0)

# Merge merged and 2023

data_merged = pd.concat([data_merged, data_2023], axis=0)

data_merged['subject_area'] = data_merged['subject_area'].str.split(', ')
data_merged = data_merged.explode('subject_area').reset_index(drop=True)

data_2018['subject_area'] = data_2018['subject_area'].str.split(', ')
data_2018 = data_2018.explode('subject_area').reset_index(drop=True)

data_2019['subject_area'] = data_2019['subject_area'].str.split(', ')
data_2019 = data_2019.explode('subject_area').reset_index(drop=True)

data_2020['subject_area'] = data_2020['subject_area'].str.split(', ')
data_2020 = data_2020.explode('subject_area').reset_index(drop=True)

data_2021['subject_area'] = data_2018['subject_area'].str.split(', ')
data_2021 = data_2021.explode('subject_area').reset_index(drop=True)

data_2022['subject_area'] = data_2018['subject_area'].str.split(', ')
data_2022 = data_2022.explode('subject_area').reset_index(drop=True)

data_2023['subject_area'] = data_2018['subject_area'].str.split(', ')
data_2023 = data_2023.explode('subject_area').reset_index(drop=True)

data_merged
```
