```python
import pandas as pd
import pybliometrics
from pybliometrics.scopus import ScopusSearch


pybliometrics.scopus.init()


from pybliometrics.scopus import SubjectClassifications
from pybliometrics.scopus import PlumXMetrics
from pybliometrics.scopus import SerialSearch
from pybliometrics.scopus import SerialTitle
```

```python
subject_code = "ENGI"  # Example: Economics
rows_to_fetch = 10000   # Total rows you want
page_size = 25         # Number of rows per request (default)

# Container for all results
all_results = []

# Loop for pagination
for start in range(0, rows_to_fetch, page_size):
    print(f"Fetching results from {start}...")
    try:
        # Fetch results
        s =  SerialSearch(query={"subj":f"{subject_code}"}, start=start, count=25, refresh=True)
        # print(s)
        if s.results:
            all_results.extend(s.results)
        else:
            break  # Stop if no more results
    except Exception as e:
        print(f"Error at start={start}: {e}")
        break

# Convert to DataFrame
df = pd.DataFrame(all_results)
```

```python
df_1 = df
# df_1 = df_1.dropna(thresh=700,axis=1)
# df.isnull().sum().to_string()
df_2 = df_1.drop(columns=['homepage','coverimage','url','scopus-source','issn','source-id','subject_area_codes','eIssn','openaccess','openaccessArticle'],axis=1)

df_2['is_ENGI'] = df_2['subject_area_abbrevs'].apply(lambda x: 1 if str(x).find("ENGI") !=-1 else 0)

df_3 = df_2[df_2['is_ENGI']==1]

cols_to_drop = [
    col for col in df.columns 
    if col.startswith(('publicationCount_','zeroCitesSCE_', 'citeCountSCE_', 'revPercent_','citeScoreCurrentMetric_')) 
    and not any(year in col for year in ['2022', '2023', '2024'])
]
cols_2_to_drop = [
    col for col in df.columns
    if col.startswith('zeroCitesPercentSCE_')
]
# print(cols_2_to_drop)
# Drop the unwanted columns
df_4 = df_3.drop(columns=cols_to_drop)
df_4 = df_4.drop(columns=cols_2_to_drop)
df_5 = df_4.drop(columns=['coverageStartYear','coverageEndYear'])
df_5 = df_5.rename(columns={'SJR_[\'@year\']':'SJR','SNIP_[\'@year\']':'SNIP'})

df_5.isnull().sum()
```

```python
df_6 = df_5.dropna(subset=['SNIP','SJR','citeScoreTracker_2024','citeScoreCurrentMetric_2023'])
df_6 = df_6.drop(columns=['is_ENGI'])
columns_to_drop = [
    "citeScoreTracker_None", 
    "openaccessType", 
    "oaAllowsAuthorPaid", 
    "openaccessWindow", 
    "openaccessStartDate"
]

# Drop the specified columns
df_7 = df_6.drop(columns=columns_to_drop)
df_7.isnull().sum()
df_7 = df_7.drop(columns=['citeScoreCurrentMetric_2022'])
df_7.isnull().sum()
```

```python
df_7 = df_7.drop(columns=['citeScoreCurrentMetric_2013','citeScoreCurrentMetric_2014'])
```
